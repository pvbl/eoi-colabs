```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python

# 1. CREAR DATOS SINTÉTICOS

X, y_true = make_blobs(
    n_samples=500,
    centers=3,
    cluster_std=0.8,
    random_state=42
)

```

```python

# 2.Inicializamos kmeans
k = 3  # número de clusters
kmeans = KMeans(
    n_clusters=k,
    random_state=42,
    n_init=10
)
```

```python

# 3. entrenamos el modelo
kmeans.fit(X)
```

```python
# 4. evaluamos resultados
labels = kmeans.labels_          # etiqueta
centroids = kmeans.cluster_centers_

# a) Valores promedio: centroides de cada cluster
print("Centroides (valores promedio de cada cluster):")
for i, c in enumerate(centroids):
    print(f"Cluster {i}: {c}")

# b) Inercia (suma de distancias cuadráticas a los centroides)
print("\nInercia (SSE):", kmeans.inertia_)
```

```python
# visualización
plt.figure(figsize=(8, 6))

# Puntos coloreados según el cluster asignado
plt.scatter(X[:, 0], X[:, 1], c=labels, s=30, alpha=0.7)

# Centroides en grande
plt.scatter(
    centroids[:, 0], centroids[:, 1],
    s=200,
    marker="X",
    edgecolor="black"
)

plt.title(f"K-Means con k={k}\nInercia={kmeans.inertia_:.2f}")
plt.xlabel("Característica 1")
plt.ylabel("Característica 2")
plt.grid(True, alpha=0.2)
plt.tight_layout()
plt.show()
```
