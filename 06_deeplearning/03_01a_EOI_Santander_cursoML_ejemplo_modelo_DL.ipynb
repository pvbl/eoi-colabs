````python


# Generar datos sintéticos
np.random.seed(0)
X_train = np.random.random((1000, 10))  # 1000 muestras, 10 características
Y_train = to_categorical(np.random.randint(4, size=(1000, 1)), num_classes=4)  # 1000 etiquetas para 4 clases

X_test = np.random.random((200, 10))  # 200 muestras, 10 características
Y_test = to_categorical(np.random.randint(4, size=(200, 1)), num_classes=4)  # 200 etiquetas para 4 clases
````

````python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.regularizers import l2

def build_optimized_model():
    model = Sequential()
    # Capa de entrada con regularización L2
    model.add(Dense(10, activation='relu', input_shape=(10,), kernel_regularizer=l2(0.001)))
    model.add(Dropout(0.3))  # Dropout para reducir sobreajuste

    # Capas ocultas con regularización L2 y Dropout
    # añade dropout y regulación
    # TODO


    # Capa de salida sin Dropout
    model.add(Dense(4, activation='softmax'))

    # Compilar el modelo con el optimizador Adam y una tasa de aprendizaje ajustada
    model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Crear el modelo optimizado
optimized_model = build_optimized_model()

# Entrenar el modelo optimizado
optimized_model.fit(X_train, Y_train, epochs=20, batch_size=32)

# Evaluar el modelo optimizado
loss, accuracy = optimized_model.evaluate(X_test, Y_test)
print(f"Optimized Test Loss: {loss:.3f}, Optimized Test Accuracy: {accuracy:.3f}")


````
